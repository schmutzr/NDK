### Seemingly Simple Communications 2/3
(vereinfacht)

* Browser benutzt _DNS_ um den _Host_/_Domainnamen_ `csail.mit.edu` in eine IP-Adresse _aufzulösen_. Da die Kommunikation im Internet _immer_ per IP-Adresse funktioniert (IPv4 oder IPv6). Bei Verwendung eines Proxies (FHNW) wird dieser Schritt vom Proxy erledigt

* Es wird eine _TCP-Verbindung_ zu der gefundenen _IP-Adresse_ aufgebaut (`128.30.2.109`). Dazu werden einzelne _IP-Pakete_ vom *TCP/IP-Stack* des Betriebssystems generiert

* Die einzelnen _IP-Pakete_ werden in _Ethernet-Frames_ eingepackt (Betriebssystem). Die Frames werden dem NIC-Treiber übermittelt, dieser sendet die Information an die _Ethernet_/_WLAN_-Hardware (_NIC_).

* das NIC verwandelt den Bitstrom in einen _Line-Code_ und erzeugt ein _Line-Signal_ (Ethernet im Baseband, WLAN zusätzlich moduliert)

* Bei Verwendung von _TLS_ (Verschlüsselung) wird zusätzlich eine kryptographisch sichere Verbindung mit dem Server ausgehandelt

* es wird ein _HTTP-Request_ gesendet:
   ```
   GET / HTTP/1.1
   Host: csail.mit.edu
   ...
   ```
   
* der Server antwortet mit dem Inhalt der Seite als _HTML_:
   ```
   200 OK
   ...
   
   <HTML...
   ```
 
* der Browser _rendert_ (visualisiert) die empfangene Information, je nach Inhalt der Seite (Bilder, etc) werden zusätzliche Requests und eventuell Verbindungen aufgebaut um alle Resourcen zu erhalten
            


### Seemingly Simple Communications 3/3

* eine Menge Akronyme und Fachbegriffe
* komplexer Ablauf mit verschiedenen Spezialgebieten (Applikationsprotokoll, Dienstprotokoll, sicherer Datenstrom, Paket-Routing, elektrisches Signal, etc)

#### Problem:
* (zu) komplexer Ablauf und Funktionen aus diversen Spezialgebieten für eine "monolitsche" Lösung (all-in-one Programm)
* eine monolitische Lösung wäre sehr unflexibel gegenüber Änderungen/Erweiterungen auf den verschiedenen Ebenen

#### Lösung:
* Aufteilung des Problems (divide-and-conquer): "Layer"-Ansatz (Schichtenmodell)
* die einzelnen Schichten kümmern sich nur um ein partikuläres Problem:
   * Darstellung des Inhalts (HTML)
   * Applikations-Kommunikation (HTTP)
   * Verschlüsselung/Autentisierung (TLS)
   * Datenstrom (TCP)
   * Paket-Wegleitung Client<->Server (IP)
   * lokale Kommunikation (L2-MAC Ethernet)
   * Umsetzung in ein elektromagnetisch (galvanisch oder über Radio/Licht) übertragbares Signal (L1-PHY Ethernet).

* jede Schicht hat genau definierte Aufgaben (Services) und ein eigenes Protokoll (Meta-Information/Overhead)
* jede Schicht kann (theoretisch) durch eine andere mit genau denselben Services (z.B. send, receive) ausgetauscht werden. Beispiel: (wired) Ethernet -> (wireless) WLAN Ethernet, die Funktion des Gesamt-Stapels wird dabei nicht tangiert

## Protocol Layers Geschichtliches

* ISO-OSI 7-Layer-Modell wurde ~1970+ als universelle Kommunikations-Abstraktion entwickelt -- Hauptanwendung  traditionelle Telekommunikation
   * als theoretisches Modell sehr informativ
   * praktische Anwendung (mittlerweile selten)
   * "Bürokraten" ("Telefönler")


* das (IEEE) TCP/IP "Internet" 5-Layer-Modell (4+) wurde pragmatisch entwickelt, der Prozess zur Verfeinerung steht theoretisch allen offen (RFC)
   * die Layer 5 bis 7 werden in eine "Applikationsschicht" zusammengefasst (normalerweise trotzdem als "Layer-7" bezeichnet)
   * vereinzelt tatsächliche 7-Layer Implementation (Sun-RPC, XDR, YP, NFS)
   * alle populären Anwendungen orientieren sich mehrheitlich am 5-Layer (1, 2, 3, 4, 7) Modell: DNS, HTTP, SMTP, IMAP, etc

## PSTN to Internet 1/5

* PSTN (Public-Switched-Telephone-Network) war in den 70' Jahren ein _circuit switched_ Netzwerk:
   * Beim Verbindungsaufbau wurde mit jeder weiter gewählten Ziffer ein weiterer 1:10 Relaisverteiler angesprochen
     z.B. [Rotary Switch](https://www.youtube.com/watch?v=xZePwin92cI)
   * jeder der 10 Ausgänge ist mit einer elektrischen Leitung zu einem nächsten Eingang verbunden.
     Die Switches sind (in Gruppen) örtlich verteilt, d.h. jede Leitung führt näher zum Empfänger

Ein solches Netzwerk ist _circuit switched_ , d.h. nach dem Verbindungsaufbau existiert eine _dedizierte_ Leitung (über die Relais) vom Anrufer zum Empfänger. Diese Leitung ist _exklusiv_ und kann während der Dauer des Anrufs nicht weiter verwendet werden.

Anmerkung: heute wird für das Telefonnetz manchmal noch der Begriff PSTN benutzt, die Technik wurde aber schon in den 80' Jahren auf _packet-switched_ umgestellt (s.u.)

### PSTN to Internet 2/5

Vor- und Nachteile der circuit-switched Technik:

$+$ exklusive Nutzung der Leitung während des Gesprächs, sehr gute Qualität

$-$ schlechte Auslastung des Netzwerks: es muss eine "Überprovisionierung" stattfinden, d.h. es müssen viel mehr Leitungen als normalerweise im Gebrauch gelegt werden

$-$ Fehleranfälligkeit: bei Ausfall eines Relais wird die Verbindung unterbrochen (und kann möglicherweise nicht neu aufgebaut werden)

$-$ "Skaliert" sehr schlecht, d.h. am Schluss sieht es dann so: [Stockholm Telephone Exchange](https://www.ericsson.com/en/blog/2014/10/when-the-telephone-lines-darkened-the-sun-in-stockholm) aus

### PSTN to Internet 3/5

Mitte der 60' Jahre traten die DARPA und das DoD (beides dem Militär nahestehende Organisationen) als Sponsor für ein _Datenkommunikationsnetzwerk_ auf, das folgende Punkte
erfüllen sollte:

* Wiederstandsfähig gegenüber Ausfällen einzelner Verbindungsknoten (resilience), auch im "Katastrophenfall"
* Skalierbar in der Grösse des Netzwerks aber auch in der verfügbaren Datenrate (scalability)
* Ökonomisch betreibbar, keine riesige Überprovisionierung wie beim (circuit-switched) PSTN

Diese Ziele waren nur erreichbar durch den Einsatz der "Packet-Switched" Technik...

Daraus entstand das [ARPANET](https://en.wikipedia.org/wiki/ARPANET), der Vorläufer des jetzigen Internets.
Das ARPANET benutzte als erstes grösseres Netzwerk die Protokollfamilie TCP/IP.

Das Design orientiert sich am Schichtenmodell: auf Frame- (L2) und Paketebene (L3) wird z.B. nicht versucht ein verlorenes Paket neu zuzustellen (keine "Quittung"). Diese Funktion wird völlig unabhängig in der Schicht L4 (TCP) implementiert.

### PSTN to Internet 4/5

Vor- und Nachteile der packet-switched Technik (Internet und nicht mehr eingesetzte ATM-Technik hauptsächlich für Telefonie):

$+$ (resilience) Jedes Datenpaket wird bei jeder Relaisstation (Router) individuell weitergeleitet. Wenn eine Strecke/Router ausfällt oder eine Überlastsituation eintritt, wird einfach der nächst beste Pfad (next-hop) bestimmt.

$+$ (scalability) die Verbindungsleitungen werden nicht mehr "exklusiv" geschaltet, d.h. Pakete mit verschiedensten Absende- und Empfängeradresse teilen sich eine Leitung ("time-domain/division-multiplexing"). Zudem können "Trunk"-Leitungen (z.B. Unterseekabel) einfach mit einer höheren Datenrate skaliert werden (anstatt mehr Leitungen zu ziehen)

$+$ (ökonomisch) durch die nicht-exlusive Nutzung der Leitungen sind auch viel weniger davon nötig. 

$-$ Der Datenstrom muss "digitalisiert" und paketiert werden (das ist eigentlich nicht unbedingt ein Nachteil)

$-$ Pakete können in unterschiedlicher Reihenfolge beim Empfänger eintreffen (Lösung: TCP)

$-$ Pakete können dupliziert werden oder ganz verlorengehen (Lösung: Applikation und TCP)

Anmerkung: in heutigen Hochleistungsnetzwerken wird oft eine Technik mit "vorkalkulierter" Route verwendet (Route-Caches und Techniken wie z.B. MPLS, Label-Switching). Der Grundgedanke des packet-switching bleibt aber auch da bestehen.

Historisch:

* Ende der 70' bis Ende 80' Jahre wurde das "Internet" von der NSF (National-Science-Foundation) betrieben.
  Der Wandel vom "militärischen" zum "wissenschaftlichen" Netzwerk war damit vollzogen.
  Viele internationale Universitäten hatten bereits einen Anschluss an das NSF-Netz.

* Anfang der 90' Jahre wurde das "Internet" kommerzialisiert, anstelle des NSF-Backbones wurden die Transit- und Endbenutzeranschlüsse von privaten Organisationen implementiert. 1991 wurde das NSF-Netz ("ARPA-Net") dekommisioniert, die ```10.0.0.0/8```-Adressen, die sie vielleicht schon gesehen haben wurden damals freigegeben
